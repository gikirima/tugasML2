# Load dataset

```python
import pandas as pd

df = pd.read_csv('transaction_data.csv', index_col=0)
df
```

```python
# transformasi dataset menjadi bentuk boolean
df = df.astype(bool)
# mengubah nilai True menjadi 1 dan False menjadi 0
df = df.astype(int)
df
```

# Penerapan Apriori

```python
import sys
!{sys.executable} -m pip install mlxtend
```

```python
# apriori algorithm
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(df, min_support=0.33, use_colnames=True) #2 barang dari 6 transaksi, sehingga support minimal 0.33
# menampilkan frequent itemsets
frequent_itemsets
```

```python
# rule association
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8) # aturan dengan confidence minimal 80%
# menampilkan rules
rules = rules[['antecedents', 'consequents', 'support', 'confidence']]
rules
```

```python
# Filter rules yang melibatkan tepat 3 barang (gabungan antecedent dan consequent)
rules_3items = rules[rules['antecedents'].apply(len) + rules['consequents'].apply(len) == 3]
rules_3items.reset_index(drop=True, inplace=True)
rules_3items
```

### Interpretasi Hasil

Hasil yang diperoleh menggunakan library `apriori` dan `association_rules` dari `mlxtend` menunjukkan frequent itemsets dan aturan asosiasi yang identik dengan hasil yang didapat melalui perhitungan manual. Hal ini dapat dilihat dari nilai support, confidence, serta kombinasi antecedents dan consequents yang sama pada kedua metode.

Dengan demikian, hasil yang diperoleh pada langkah manual telah tervalidasi dan terbukti benar, karena sesuai dengan hasil dari library yang sudah teruji secara luas. Ini menunjukkan bahwa proses perhitungan manual yang dilakukan sudah tepat dan dapat diandalkan.
