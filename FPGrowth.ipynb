```python
import pandas as pd

df = pd.read_csv('transaction_data.csv', index_col=0)
df
```

```python
# ubah menjadi boolean, kemudian ubah menjadi integer
df = df.astype(bool).astype(int)
df
```

```python
# Penerapan Algoritma FP Growth
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules

# buat frequent itemset dengan fpgrowth
frequent_itemsets = fpgrowth(df, min_support=0.33, use_colnames=True)

# tampilkan hasil
frequent_itemsets
```

```python
# membangun association rules dengan minimum confidence 80%
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.8)

# tampilkan hasil anteseden, konsekuen, support, dan confidence
rules[["antecedents", "consequents", "support", "confidence"]]
```
